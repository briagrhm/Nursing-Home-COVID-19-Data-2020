```python
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
# Read the data into a Pandas DataFrame
df = pd.read_csv("faclevel_2020.csv",low_memory=False)
# Get the sheet names.
df.head()
```

```python
# Drop the empty rows
df_drop=df.drop(df.columns[[23,24,25,26,27,28,29,30,31]], axis=1)
df_drop
```

```python
df_drop.info()
```

```python
#Rename the long column Names 
df_rename = df_drop.rename(columns={'Residents Hospitalizations with Confirmed COVID-19':'Hospitalizations with Confirmed COVID 19 ','Residents Hospitalizations with Confirmed COVID-19 and Up to Date with Vaccines':'Hospitalizations Confirmed COVID 19 and Up to Date Vaccines'})
```

```python
#Change data type for the date 
df_rename['Week Ending']= pd.to_datetime(df_rename['Week Ending'], format='%m/%d/%y')
df_rename['Week Ending'].info()
```

```python
#check for missing data 
df_rename.isnull().sum()
```

```python
#Drop the rows with missing data 
df_clean=df_rename.dropna()
df_clean.isnull().sum()
```

```python
# Resize the figure
plt.figure(figsize=(20, 6))
#Create the plot 
plt.plot(df_rename["Provider State"], df_rename["Residents Total COVID-19 Deaths"])
plt.xlabel("Provider States")
plt.ylabel("Residents Total COVID-19 Deaths")
plt.title("Nursing Home Resident Death By State 2020")
# Rotate x-axis labels for better visibility
plt.xticks(rotation=45)
plt.show()
```

```python
# Resize the figure
plt.figure(figsize=(20, 6))
#Create the plot 
plt.bar(df_rename["Provider State"], df_rename["Residents Total COVID-19 Deaths"])
plt.xlabel("Provider States")
plt.ylabel("Residents Total COVID-19 Deaths")
plt.title("Nursing Home Resident Death By State 2020")
# Rotate x-axis labels for better visibility
plt.xticks(rotation=45)
plt.show()
```
